```python
class A:
    def saludar(self):
        print("Hola desde A")

class B(A):
    def saludar(self):
        print("Antes")
        super().saludar()
        print("DespuÃ©s")

# Fix: Create an instance of B and call the method on the instance
instance_b = B()
instance_b.saludar()
print(B)
```

```python
class Base:
    def __init__(self):
        print("Base")

class X(Base):
    def __init__(self):
        super().__init__()
        print("X")

class Y(Base):
    def __init__(self):
        super().__init__()
        print("Y")

class Z(X, Y):
    def __init__(self):
        super().__init__()  # sigue el MRO y llama a X -> Y -> Base
        print("Z")
instance_Z = Z()
print(instance_Z)
```

```python
class Auto:
    def __init__(self, color: str, velocidad_actual: float=0, multiplicador: float=1):
        self.color = color
        self.velocidad_actual = velocidad_actual
        self.multiplicador = multiplicador
    def acelerar(self, aceleracion: float, tiempo: float) ->float:
        self.velocidad_actual += aceleracion*tiempo*self.multiplicador
        return f"La velocidad es esta = {self.velocidad_actual} m/s"
    def frenar(self, desaceleracion: float) ->float:
        self.velocidad_actual = max(0, self.velocidad_actual - desaceleracion)
        if self.velocidad_actual == 0:
            return "El auto se ha detenido por completo"
        return f"La velocidad actual es:{self.velocidad_actual:.2f} m/s"
    def mostrar_estado(self) ->str:
        return f"La velocidad actual: {self.velocidad_actual:.2f} m/s"

class Deportivo(Auto):
    def __init__(self,color: str, velocidad_actual: float=0, nombre: str=None):
        super().__init__(color, velocidad_actual, multiplicador=1.5)
        self.nombre = nombre
    def turbo(self, extra:float):
        self.velocidad_actual += extra*self.multiplicador
        return f"Turbo activado: Velocidad actual: {self.velocidad_actual:.2f} m/s"

def carro_accion():
    Ferrari = Deportivo("Rojo", nombre="Ferrari")
    while True:
        opcion = input("Escoge accion: ").replace(" ", "").lower()
        if opcion == "acelerar":
            aceleracion = float(input("cuanto quires acelerar: ").replace(" ", "").lower())
            tiempo = float(input("por cuanto tiempo vas a acelerar: ").replace(" ", "").lower())
            print(Ferrari.acelerar(aceleracion, tiempo))
        elif opcion == "turbo":
            turbo = float(input("cuanto turbo daras: ").replace(" ", "").lower())
            print(Ferrari.turbo(turbo))
        elif opcion == "frenar":
            desaceleracion = float(input("que tanto frenaras: ").replace(" ", "").lower())
            print(Ferrari.frenar(desaceleracion))
        elif opcion == "mostrar":
            print(Ferrari.mostrar_estado())
        elif opcion == "terminar":
            print("saliendo del programa...")
            break
        else:
            print("Escoge la opcion correcta")
carro_accion()
```
