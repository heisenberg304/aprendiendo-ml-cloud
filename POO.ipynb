```python
class Auto:
    def __init__(self, marca, color, velocidad_actual):
        self.marca = marca
        self.color = color
        self.velocidad_actual = velocidad_actual
        self.multiplicador = 1

    def acelerar(self):
        try:
            aceleracion = float(input("Escoga la aceleracion: "))
            tiempo = float(input("Tiempo que se desplazara (en segundos): "))
            self.velocidad_actual += aceleracion*tiempo*self.multiplicador
            return f"La velocidad es esta = {self.velocidad_actual} m/s"
        except ValueError:
            return "Valor no valido"
    def frenar(self):
        try:
            desaceleracion = float(input("Ingrese el grado de desaceleracion: "))
            if desaceleracion > self.velocidad_actual:
                self.velocidad_actual = 0
                return "El auto se ha detenido por completo"
            else:
                self.velocidad_actual -= desaceleracion
                return self.velocidad_actual
        except ValueError:
            return "Valor no valido"
    def mostrar_estado(self):
        return f"La velocidad actual es esta = {self.velocidad_actual} m/s"
    def actualizar_multiplicador(self, nuevo):
        self.multiplicador = nuevo
        return self.multiplicador

def carro_accion(carro):
    while True:
        opcion = input("Escoge accion: ").replace(" ", "").lower()
        if opcion == "acelerar":
            print(carro.acelerar())
        elif opcion == "frenar":
            print(carro.frenar())
        elif opcion == "mostrar":
            print(carro.mostrar_estado())
        elif opcion == "terminar":
            print("saliendo del programa...")
            break
        else:
            print("Escoge la opcion correcta")
carro1 = Auto("Toyota", "Amarillo", 0)

carro2 = Auto("Ferrari", "negro", 0)
carro2.actualizar_multiplicador(1.5) #al ser ferrari sera mas rapido y por eso tiene un multiplicador mas grande

print(carro_accion(carro2))
```
